```python
movies = pd.read_csv('/content/drive/MyDrive/DSA4213/movies_full.csv')
```

```python
movies.dropna(inplace=True)
movies.reset_index(drop=True, inplace=True)
```

```python
#movies
```

```python
movies['average_rating'] = movies['average_rating'].astype(str)
movies['tags'] = movies['genres'] + ' ' + movies['tag'] + movies['average_rating']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import h2o
from h2o.automl import H2OAutoML
import pickle
```

```python
# Feature extraction with CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
def recommend(movie_title):
    movie_title = movie_title.strip()

    # Find the indices of movies with the provided title (case-insensitive)
    movie_indices = movies.index[movies['title'].str.strip().str.lower() == movie_title.lower()].tolist()

    if not movie_indices:
        print("Movie not found.")
        return
    idx = movie_indices[0]
    distances = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda x: x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]]['title'])
```

```python
# Save necessary objects with pickle
pickle.dump(movies, open('movies_list.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```

```python
recommend("Toy Story")
```
